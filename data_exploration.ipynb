# Imports

```python
import calendar
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import random
```

# Create dataframe of all year sales

```python
months = list(calendar.month_name)
dataframes = list()

for num in range(1, 13, 1):
    dataframes.append(pd.read_csv('Data/Sales_' + months[num] + '_2019.csv'))

sales_df = pd.concat(dataframes)
sales_df.to_csv('Data/Sales_2019.csv', index=False)
sales_df
```

# Generate Graphs

```python
def bar_plot(data: list, title: str, filename: str):
    
    species = ['Simple MapReduce', 'Combiner Based', 'In Mapper']
    penguin_means = {
        '1 Core': data[0],
        '10 Core': data[1]
    }

    x = np.arange(len(species))
    width = 0.25
    multiplier = 0

    fig, ax = plt.subplots(layout='constrained')

    for attribute, measurement in penguin_means.items():
        offset = width * multiplier
        rects = ax.bar(x + offset, measurement, width, label=attribute)
        ax.bar_label(rects, padding=3)
        multiplier += 1

    # Add some text for labels, title and custom x-axis tick labels, etc.
    ax.set_ylabel('Time Taken', fontsize = 15)
    ax.set_title(title, fontsize = 15)
    ax.set_xlabel('Design Patterns', fontsize = 15)
    ax.set_xticks(x + width, species)
    ax.legend(loc='upper right', ncols=3)
    plt.savefig(f'Images/{filename}')
    plt.show()
```

## Item Count

```python
title = 'Design Pattern Vs. Time Taken for Item Count'
data = [[4.3122, 4.1493, 0.8144], [1.68887, 1.52059, 0.875332]]
filename = 'item_count.png'
bar_plot(data, title, filename)
```
